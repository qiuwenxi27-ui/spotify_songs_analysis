```python
import pandas as pd
df=pd.read_csv('spotify_songs.csv')
df.head()
print(df.shape)
df.info()

```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(10,6))
sns.histplot(df['track_popularity'])
plt.title('The distribution of song popularity')
plt.show()
sns.scatterplot(data=df, x='energy', y='track_popularity', alpha=0.3)
plt.title('energy vs popularity')
plt.show()         
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(10,6))
df.groupby('playlist_genre')['valence'].mean().sort_values().plot(kind='bar')
plt.title('The nvalence in diffrent playlist_genre')
plt.ylabel('valence mean')
plt.show()   
```

```python
sns.scatterplot(data=df, x='danceability', y='energy', alpha=0.3)
plt.title('danceability vs energy')
plt.show()  
```
